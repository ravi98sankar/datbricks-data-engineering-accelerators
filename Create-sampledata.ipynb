```python
# COMMAND ----------
# STEP 5: Demonstrate Schema Evolution on purchase_history Table



from pyspark.sql.types import StructType, StructField, StringType, DoubleType, TimestampType, IntegerType
from datetime import datetime

# Add a new column 'loyalty_tier' to the schema
evolved_schema = StructType([
    StructField("user_id", StringType(), False),
    StructField("transaction_id", StringType(), False),
    StructField("category", StringType(), False),
    StructField("amount", DoubleType(), False),
    StructField("purchase_date", TimestampType(), False),
    StructField("loyalty_tier", StringType(), True)
])

# New data with the evolved schema
evolved_data = [
    ("1004", "TXN_014", "electronics", 500.00, datetime(2026, 5, 15, 10, 0), "gold"),
    ("1005", "TXN_015", "apparel", 75.00, datetime(2026, 5, 16, 12, 30), "silver")
]

df_evolved = spark.createDataFrame(evolved_data, schema=evolved_schema)

# Write with schema evolution (mergeSchema option)
df_evolved.write \
    .format("delta") \
    .mode("append") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{CATALOG_NAME}.{SCHEMA_NAME}.{TABLE_NAME}")

print("Schema evolution: Added 'loyalty_tier' column.")

# Show the updated schema
display(
    spark.sql(f"DESCRIBE TABLE EXTENDED {CATALOG_NAME}.{SCHEMA_NAME}.{TABLE_NAME}")
)

# Show the data including the new column
display(
    spark.sql(f"SELECT * FROM {CATALOG_NAME}.{SCHEMA_NAME}.{TABLE_NAME} ORDER BY purchase_date DESC")
)
```
